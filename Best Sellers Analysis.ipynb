## Data Analysis

1. Are the Number of Reviews increasing each year?
2. Which book received the most number of views?

### Q1. Are the Number of Reviews increasing each year?

The number of reviews has seen increasing over the years especially the years 2018 to 2019 has garnered 450,000 thousand reviews in those two years.
A downward trend happen between 2010 to 2011 shifting form 200,000 reviews in 2010 down to only 145,000 thousand in 2011. However the slump is only that year as the succeeding years we've seen an increase by two-folds 326,000 thousand in 2012.

```python
review = df.groupby('Year', as_index = False).agg({'Reviews':'sum'})
```

```python
sns.lineplot(x='Year', y='Reviews', data = review, marker='o')

plt.xlabel('')
plt.ylabel('Number of Reviews')
plt.show()
```

### Q2. Which book received the most number of views?

The most reviewed books is "Where the Crawdads Sing" by Delia Owens which garnered 87,000  reviews

```python
df = df.sort_values(by = 'Reviews', ascending = False)
books = df.head(10)
books
```

```python
sns.barplot(y='Name', x = 'Reviews', data = books, palette = 'pastel')
plt.title('Most Reviewed Books')
plt.ylabel('')
plt.xlabel('')
plt.show()
```
